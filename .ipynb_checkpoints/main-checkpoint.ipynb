# CASE STUDY : ANALYSIS OF RIDE BOOKING DATASET  

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
 
# read CSV file
bd=pd.read_csv(r"C:\Ride_Booking_Analysis\Dataset.csv")
 
#Shape of the dataset
display( bd.shape)
display(bd.head(10))
 
#columns
display(list(bd.columns))
 
```

```python
#Check for unique rows
bd.nunique()
#Booking ID and Customer ID are supposed to be the unique attributes

#bd=bd.drop_duplicates(subset="Booking ID")
#bd=bd.drop_duplicates(subset="Customer ID")
display(" SHAPE ",bd.shape)


#Function to check not cancelled entries to avoid filling null values for cancelled rows
def not_cancelled(df):
    return df["Booking Status"].isin(["Cancelled by Driver", "Cancelled by Customer"]) == False
```

## --------------------------------------------------------------------------------------------------------------------------------------

### 1.	Data Cleaning & Preprocessing.

#### 1) Handle missing or inconsistent data.

```python
#SUM of NULL values

display(bd["Booking Status"].value_counts())
bd.isnull().sum().reset_index()
 
```

```python
#Avg VTAT
#check of outliers

#display(bd["Avg VTAT"].plot(kind="box"))

#Replace null values with appropriate 

bd.boxplot(column="Avg VTAT", boxprops=dict(color="black"))
plt.show()
 

bd.loc[not_cancelled(bd), "Avg VTAT"] = bd.loc[not_cancelled(bd), "Avg VTAT"].fillna(bd["Avg VTAT"].mean())
    
bd["Avg VTAT"]=round(bd["Avg VTAT"],2) 
```

```python
#Avg CTAT

bd.boxplot(column="Avg CTAT", boxprops=dict(color="black"))
plt.show()


bd.loc[not_cancelled(bd), "Avg CTAT"] = bd.loc[not_cancelled(bd), "Avg CTAT"].fillna(bd["Avg CTAT"].mean())

bd["Avg CTAT"]=round(bd["Avg CTAT"],2)

display(bd.head())
 
```

```python
#Pyment method 
bd["Payment Method"].value_counts()

#since its a categorical value
    
bd.loc[not_cancelled(bd), "Payment Method"] = bd.loc[not_cancelled(bd), "Payment Method"].fillna(bd["Payment Method"].mode()[0])


display(bd.head(10))
```

```python
bd["Customer Rating"].value_counts()

    
bd.loc[not_cancelled(bd), "Customer Rating"] = bd.loc[not_cancelled(bd), "Customer Rating"].fillna(bd["Customer Rating"].mean())
bd.loc[not_cancelled(bd), "Driver Ratings"] = bd.loc[not_cancelled(bd), "Driver Ratings"].fillna(bd["Driver Ratings"].mean())

bd["Customer Rating"]=round(bd["Customer Rating"],1)
bd["Driver Ratings"]=round(bd["Driver Ratings"],1)

 
    
display(bd.head())
```

```python
bd["Incomplete Rides"].value_counts()
 
bd["Booking ID"].nunique()
```

```python
#Bookig Values

#Check for outliers
#bd["Booking Value"].plot(kind="box")
display(bd.boxplot(column="Booking Value", boxprops=dict(color="black")))
 
bd.loc[not_cancelled(bd), "Booking Value"] = bd.loc[not_cancelled(bd), "Booking Value"].fillna(bd["Booking Value"].median())

display(bd.head())
```

```python
#Ride Distance
#Outliers
#display(bd["Ride Distance"].plot(kind="box"))

display(bd.boxplot(column="Ride Distance", boxprops=dict(color="black")))

bd.loc[not_cancelled(bd), "Ride Distance"] = bd.loc[not_cancelled(bd), "Ride Distance"].fillna(bd["Ride Distance"].mean())


bd["Ride Distance"]=round(bd["Ride Distance"],2)
```

```python
#SUM of NULL values
display(bd.isnull().sum().reset_index())
display(bd.head())
 
```

#### Convert date/time columns to appropriate formats.

```python
#DATE Formatting

# Fix date formatting
bd["Date"] = pd.to_datetime(bd["Date"], format="%m/%d/%Y", errors="coerce")
bd["Date"] = bd["Date"].dt.strftime("%d/%m/%Y")

# Sort and reset index correctly
bd = bd.sort_values(by="Date")
bd = bd.reset_index(drop=True)


#Time Formatting

#bd["Time"]=pd.to_datetime(bd['Time'],format="%H:%M:%S")

bd["Time"] = pd.to_datetime(bd["Time"], format="mixed", errors="coerce")


bd["Time"] = bd["Time"].dt.strftime("%I:%M:%S%p")



# Display
display(bd.head(20))



 
```

```python
#OUTPUT

bd.to_csv("output.csv",index=False)
```
